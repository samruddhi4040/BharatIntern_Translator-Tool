# Access the Google translate functionality for language translation & interactive widgets in Jupyter notebooks. 

```python
from googletrans import Translator
from ipywidgets import interact, widgets
from IPython.display import display
```

# Taking user-input & producing the output

```python
def translatorapp():
    translate_from_dropdown = widgets.Dropdown(
        options=['English', 'Hindi', 'Marathi'],
        description='Translate from:',
    )

    translate_to_dropdown = widgets.Dropdown(
        options=['English', 'Hindi', 'Marathi'],
        description='Translate to:',
    )

    text_input = widgets.Textarea(
        description='Enter text to be translated:'
    )

    output_text = widgets.Output()

    def update_translation(*args):
        with output_text:
            output_text.clear_output()
            translate_from = translate_from_dropdown.value.lower()
            translate_to = translate_to_dropdown.value.lower()
            text = text_input.value.strip()

            if text:
                translator = Translator()
                translated_text = translator.translate(text, src=translate_from, dest=translate_to)
                print(f"Translated Text: {translated_text.text}")
            else:
                print("Enter text to translate.")

    translate_from_dropdown.observe(update_translation, 'value')
    translate_to_dropdown.observe(update_translation, 'value')
    text_input.observe(update_translation, 'value')

    display(widgets.VBox([translate_from_dropdown, translate_to_dropdown, text_input, output_text]))
```

# Calling the function 

```python
translatorapp()
```
